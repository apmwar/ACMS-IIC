```python
import pandas as pd
import nltk
import csv

stop_words = nltk.corpus.stopwords.words('english')

newStopWords = ['something','Im','I','please','thank you','thanks']
stop_words.extend(newStopWords)

user_info=pd.read_csv('Amazon SNS.tsv',delimiter='\t',encoding='utf-8')
with open('sample.tsv', 'w', encoding='utf-8') as tsvfile:
           writer = csv.writer(tsvfile, delimiter='\t')
           for i in range(len(user_info)):
               word_tokens = nltk.word_tokenize(user_info.values[i][1])
               print(word_tokens)
               filtered_sentence = [w for w in word_tokens if not w in stop_words] 
               filtered_sentence = [] 
               for w in word_tokens: 
                   if w not in stop_words:
                       filtered_sentence.append(w)
               writer.writerow(['Amazon EC2', filtered_sentence])
               # print(str(filtered_sentence))
```
